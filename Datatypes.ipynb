## Sequences
- list
- tuple
- set
- dictionary


```python
list_data_1 = [1,2,3]  #list of int items
list_data_2 = [2.3,4.5,6.7]  #list of float items
list_data_3 = ["A", "Saturday","Blue"] #list of text items
tuple_data = (1,2,3) #tuple created with help of ()
set_data ={1,2,3} #set created with help of {}
```

```python
print(list_data_1)
print(list_data_2)
print(list_data_3)
print(tuple_data)
print(set_data)
```

### Accessing an item from list

- index is an indicator to point to a specific element in the list.
- Index starts from 0
- Set variable cannot be subscriptable with an index

```python
list_data_1[0], list_data_2[1], list_data_3[2]
```

```python
tuple_data[1]
```

```python
set_data[1]
```

Dictionary

```python
dict_data = {"A":1, "B":2, "C":3}
print(dict_data)
```

### Accessing an item from dictionary

- Key is an indicator to point to a specific element in the dictionary.
- Every key has a value

```python
dict_data["B"]
```
